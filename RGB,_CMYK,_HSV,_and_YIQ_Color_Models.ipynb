# **RGB to CMYK using Pillow (PIL):**

```python
import cv2
import matplotlib.pyplot as plt

# Load the RGB image
rgb_image = cv2.imread('/content/cv_image.jpg')

# Display the RGB image
plt.subplot(121)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
plt.axis('off')

# Create a simulated CMYK image (without K channel)
cmyk_simulated = rgb_image.copy()
cmyk_simulated[:, :, 0] = 255 - cmyk_simulated[:, :, 0]  # Simulated Cyan
cmyk_simulated[:, :, 1] = 255 - cmyk_simulated[:, :, 1]  # Simulated Magenta
cmyk_simulated[:, :, 2] = 255 - cmyk_simulated[:, :, 2]  # Simulated Yellow

# Display the simulated CMYK image
plt.subplot(122)
plt.imshow(cmyk_simulated)
plt.title('Simulated CMYK Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **RGB to HSV using OpenCV:**


```python
import cv2
import matplotlib.pyplot as plt

# Load the RGB image
rgb_image = cv2.imread('/content/cv_image.jpg')

# Convert to HSV
hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2HSV)

# Display the RGB image
plt.subplot(121)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
plt.axis('off')

# Display the HSV image
plt.subplot(122)
plt.imshow(hsv_image)
plt.title('HSV Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **RGB to YIQ using Pillow (PIL):**

```python
import cv2
import matplotlib.pyplot as plt

# Load the RGB image
rgb_image = cv2.imread('/content/cv_image.jpg')

# Convert to YIQ
def rgb_to_yiq(rgb_image):
    # Define the transformation matrix
    yiq_transform = np.array([[0.299, 0.587, 0.114],
                              [0.596, -0.274, -0.322],
                              [0.211, -0.523, 0.312]])

    # Reshape the image to a 2D array
    pixels = rgb_image.reshape((-1, 3)).T

    # Convert to YIQ
    yiq_image = np.dot(yiq_transform, pixels).T
    yiq_image = yiq_image.reshape(rgb_image.shape)

    return yiq_image

yiq_image = rgb_to_yiq(rgb_image)

# Display the RGB image
plt.subplot(121)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
plt.axis('off')

# Display the YIQ image
plt.subplot(122)
plt.imshow(yiq_image.astype(np.uint8))
plt.title('YIQ Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
