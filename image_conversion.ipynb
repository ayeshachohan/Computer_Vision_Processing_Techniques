```python
import cv2
import matplotlib.pyplot as plt

# Load the RGB image
image = cv2.imread('/content/lena_std.tif')

# Convert to grayscale using the luminance method
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display the original RGB image in the first subplot
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')  # Hide axis for the original image

# Display the grayscale image in the second subplot
axes[1].imshow(gray_image, cmap='gray')
axes[1].set_title('Grayscale Image')
axes[1].axis('off')  # Hide axis for the grayscale image

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```

# **Method 2: Use imread()**

```python
import cv2
import matplotlib.pyplot as plt

# Read the grayscale image using imread()
gray_image = cv2.imread('/content/lena_std_gray.tif', cv2.IMREAD_GRAYSCALE)

# Read the RGB image using imread()
rgb_image = cv2.imread('/content/lena_std_rgb.tif')

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide axis
plt.show()

# Display the RGB image
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
plt.axis('off')  # Hide axis
plt.show()
```

# **Method 3: Use NumPy**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the grayscale image using OpenCV
gray_image = cv2.imread('/content/lena_std_gray.tif', cv2.IMREAD_GRAYSCALE)

# Read the RGB image using OpenCV
rgb_image = cv2.imread('/content/lena_std_rgb.tif')

# Display the grayscale image using NumPy
plt.imshow(np.asarray(gray_image), cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide axis
plt.show()

# Display the RGB image using NumPy
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')
plt.axis('off')  # Hide axis
plt.show()
```

# **Method 4: Use Matplotlib and Sci-Kit-Learn**

```python
from skimage import io
import matplotlib.pyplot as plt

# Read the grayscale image using skimage
gray_image = io.imread('/content/lena_std_gray.tif', as_gray=True)

# Read the RGB image using skimage
rgb_image = io.imread('/content/lena_std_rgb.tif')

# Display the grayscale image using Matplotlib
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide axis
plt.show()

# Display the RGB image using Matplotlib
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.axis('off')  # Hide axis
plt.show()
```
